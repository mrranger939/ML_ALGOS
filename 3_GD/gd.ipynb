```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1,n_informative=1,n_targets=1,noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
# lets apply OLS first
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
X
```

```python
reg.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red')
""" You're passing 4 points for X and 4 points for reg.predict(X) — 
so matplotlib just draws straight lines connecting those 4 points in order.
It does not "fit" a full smooth line like a continuous curve — instead, 
it connects the predicted points one by one, like this: """
```

```python
y_pred = ((78.35*X) + 0).reshape(4) # asumming b = 0
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='blue', label='y_pred')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
loss_shape = -2*np.sum(y-m*X.ravel()-b) # ravels flattens the array
loss_shape
```

```python
lr = 0.1
step_size = loss_shape*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred = ((78.35*X) + b).reshape(4) 
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='blue', label='y_pred')
plt.legend()
plt.show()
```

```python
for _ in range(5):
    b = b - (-2*np.sum(y-m*X.ravel()-b))*lr
    y_pred = ((78.35*X) + b).reshape(4) 
    plt.scatter(X,y)
    plt.plot(X, reg.predict(X), color='red', label='OLS')
    plt.plot(X, y_pred, color='blue', label='y_pred')
    plt.legend()
    plt.show() 

```
